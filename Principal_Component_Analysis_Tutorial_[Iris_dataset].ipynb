#PCA 
is a machine learning model that reduces the dimension of the feature space, in which new variables are orthogonal and linearly independent to each other.

Goal of PCA: determine linearly independent dimensions that represent the data points. --------Rishav Kumar 
(https://medium.com/@aptrishu/understanding-principle-component-analysis-e32be0253ef0)

```
# Code source: GaÃ«l Varoquaux
# Modified for documentation by Jaques Grobler
# License: BSD 3 clause

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
from sklearn.decomposition import PCA
```

```
# import iris dataset
iris = datasets.load_iris()
X = iris.data[:, :2] # we only take the first two features.
print(X)
y = iris.target
print(y)
```

```
x_min, x_max = X[:, 0].min() - 5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
```

```
plt.figure(2, figsize=(8, 6))
plt.clf()
```

```
# Plot the training points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1,
            edgecolor = 'k')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
```

```
# To getter a better understanding of interaction of the dimensions
# plot the first three PCA dimensions

fig = plt.figure(1, figsize=(8, 6))
print(fig)
ax = Axes3D(fig, elev= -150, azim=110)
print(ax)
```

```
X_reduced = PCA(n_components=3).fit_transform(iris.data)
```

```
ax.scatter(X_reduced[:,0],
           X_reduced[:,1],
           X_reduced[:,2],
           c=y,
           cmap=plt.cm.Set1,
           edgecolor='k',
           s=40)
print(ax.scatter)
```

```
ax.set_title("First three PCA directions")
ax.set_xlabel("1st elements")
ax.w_xaxis.set_ticklabels([])
```

```
ax.set_ylabel("2nd eigenvector")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd eigenvector")
ax.w_zaxis.set_ticklabels([])
```

```
plt.show()
#plt.show must be in same column with those labellling and plotting lines
```
